```python
import spacy

nlp = spacy.load('en_core_web_lg')
```

```python
doc = nlp('Dog cat banana sjkhdc')

for token in doc:
    print(token.text, "Vector :",token.has_vector, "OOV: ", token.is_oov)
    
# Spacy large model has been trained on global words ...So for words which it has not seen during training, it will not have the vector built for that or added to its vocabulary
```

```python
base_token=nlp('bread')
print(base_token[0].vector.shape)
base_token[0].vector
```

```python
doc = nlp('bread burger wheat sandwich car truck person people')

for token in doc:
    print(f"{token.text} <-> {base_token.text}:", token.similarity(base_token))
```

These models are based upon the context in which the words appear and may or may not resemble to the actual similarity which we use in day to day language
profit & loss may appear in sam econtext and may have higher similarity in these models although these 2 words are exactly opposite

```python
def print_similarity(base_word, words_to_compare):
    base_token = nlp(base_word)
    doc = nlp(words_to_compare)
    for token in doc:
        print(f"{token.text} <-> {base_token.text}:", token.similarity(base_token))
```

```python
print_similarity("iphone", "apple samsung iphone dog kitten")
```

```python
king = nlp.vocab['king'].vector
man = nlp.vocab['man'].vector
woman = nlp.vocab['woman'].vector
queen = nlp.vocab['queen'].vector

result = king -man + woman
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity([result],[queen])
```
